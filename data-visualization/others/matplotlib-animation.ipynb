```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import random
from vega_datasets import data
```

```python
%matplotlib notebook
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x + 2*np.pi)

plt.plot(x, y)
```

## Data preprocessing

```python
df = data.cars()
df.head()
```

```python
# Filter out NaN rows in a specific column
df.dropna(subset=['Horsepower', 'Miles_per_Gallon'], inplace=True)
```

```python
# Transform data
x = df['Horsepower'].to_numpy().reshape(-1, 1)
y = df['Miles_per_Gallon'].to_numpy().reshape(-1, 1)
```

```python
df.info()
```

```python
plt.scatter(x, y, label='Horsepower vs. Miles_per_Gallon')
plt.legend()
plt.show()
```

## Animation

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x_data = []
y_data = []

fig, ax = plt.subplots()
ax.set_xlim(30, 250)
ax.set_ylim(5, 50)
line, = ax.plot([], [], 'o', label='Horsepower vs. Miles_per_Gallon')
ax.legend()

def animation_frame(i):
    x_data.append(x[i])
    y_data.append(y[i])

    line.set_xdata(x_data)
    line.set_ydata(y_data)
    return line, 

animation = FuncAnimation(fig, func=animation_frame, frames=len(x), interval=10)
plt.show()
```
