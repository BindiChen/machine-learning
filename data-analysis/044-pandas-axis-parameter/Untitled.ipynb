The axis parameter is used for choosing the axis for the function to be applied. 0 is for the index and 1 is for the columns. The default value is usually 0. I have never used a function with a default axis parameter value of 1.
Consider the sum function. When we apply it to a dataframe, it calculates the sum of the values along the rows or columns.

When the axis is 0, the sum operation is done along the index so the output gives the column sums. When the axis is 1, rows sums are calculated.
The axis is a valid parameter for a lot of functions such as mean, sort_values, groupby, fillna, dropna, rank, and so on.
